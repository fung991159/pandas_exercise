```python
import pandas as pd

xls = r'O:\Payroll Services\Unit4\Batch#4\KCSC406-LENOVOHK-Lenovo\PaySummary_20191008.xlsx'

df_pt = pd.read_excel(xls, sheet_name=1)
df_ps = pd.read_excel(xls, sheet_name=0)
```

```python
for i in range(2014,2020):
    df_ps_yearly = df_ps[df_ps.PayRunCode.str.contains(str(i))]
    df_pt_yearly = df_pt[df_pt.PayRunCode.str.contains(str(i))]
    
    with pd.ExcelWriter(r'O:\Payroll Services\Unit4\Batch#4\KCSC406-LENOVOHK-Lenovo\PaySummary_{}.xlsx'.format(i)) as writer: 
        df_ps_yearly.to_excel(writer, 'PaySummary', index=False)
        df_pt_yearly.to_excel(writer, 'PayTransaction', index=False)


    
```
